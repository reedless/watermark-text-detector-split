```python
from watermark.main import get_dataset_dicts
```

```python
dataset_dicts = get_dataset_dicts("dataset/train/watermark_input", "dataset/train/watermark_mask")
```

```python
import random
import cv2
from detectron2.utils.visualizer import Visualizer
from detectron2.data import DatasetCatalog, DatasetMapper, MetadataCatalog
import matplotlib.pyplot as plt


watermarks_metadata = MetadataCatalog.get("watermarks_train")

d = [f for f in dataset_dicts if f['image_id'] == '1003.jpg'][0]
img = cv2.imread(d["file_name"])
visualizer = Visualizer(img[:, :, ::-1], metadata=watermarks_metadata, scale=0.5)
out = visualizer.draw_dataset_dict(d)
plt.imshow(out.get_image()[:, :, ::-1])
```
